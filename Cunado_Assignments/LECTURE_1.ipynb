1. Recall the definition of a rational number, denoted as $\mathbb{Q}$. Prove that the Euler's number $e = \Sigma_{k=0}^\infty \frac{1}{k!} \notin \mathbb{Q}$. A factorial is defined as $k! = (k)(k-1)(k-2)(k-3)..., \forall k \in \mathbb{Z}^+$, note that $0! = 1$. Furthermore, a sum notation $\Sigma_{k=0}^\infty k = 0+ 1 + 2 + 3 +....+...$

**ANSWER:**

Step 1: To prove that Euler's number
 is irrational, we'll proceed by contradiction.

  Assume, for the sake of contradiction, that $e$ is rational. This means that $e$ can be expressed as a fraction $\frac{p}{q}$ where $p$ and $q$ are integers with no common factors other than 1, and $q \neq 0$.

Step 2: Recall the definition of $e$ as the sum of the infinite series:


$ e = \Sigma_{k=0}^\infty \frac{1}{k!}$
​

Step 3: Consider the partial sums of the series. Let $S_n$ denote the $n$th partial sum:

$S_n = \Sigma_{k=0}^\infty \frac{1}{k!}$

Step 4: Recognize that $e$ is defined as the limit of these partial sums as $n$ approaches infinity:

$e = \lim_{n \to \infty} S_n$

Step 5: Since $e$ is rational, $S_n$ must also approach a rational limit as $n$ goes to infinity.

Step 6: Notice that each term in the series $\frac{1}{k!}$ is positive. Therefore, each partial sum $S_n$ is increasing.

Step 7: Consider the factorial terms $k!$. For any $k \geq 1$, we have $k! > k$, which implies $\frac{1}{k!} < \frac{1}{k}$.

Step 8: Recognize that the series $\sum_{k=0}^\infty \frac{1}{k}$ diverges (known as the harmonic series).

Step 9: Since $\frac{1}{k!} < \frac{1}{k}$, we can conclude that the series $\sum_{k=0}^\infty \frac{1}{k!}$ converges.

Step 10: If $S_n$ converges to a rational number, it must be bounded.

Step 11: But the convergence of $\sum_{k=0}^\infty \frac{1}{k!}$ contradicts the unboundedness of the harmonic series.

**In Conclusion**, assuming $e$ is rational leads to a contradiction, we conclude that $e$ is not rational, i.e., $e \notin \mathbb{Q}$.

2. Prove Minkowski's Inequality for sums, $\forall \ p>1, (a_k, b_k)>0$:
\begin{gather}
\begin{bmatrix}
\Sigma_{k=1}^n |a_k + b_k|^p
\end{bmatrix}^\frac{1}{p}
\leq
\begin{bmatrix}
\Sigma_{k=1}^n |a_k|^p
\end{bmatrix}^\frac{1}{p}
+
\begin{bmatrix}
\Sigma_{k=1}^n |b_k|^p
\end{bmatrix}^\frac{1}{p}
\end{gather}

**ANSWER:**

Step 1: Laws of Exponents (Power raised to a power):

- Let $n = $ any base number, then ${(n^p)}^\frac{1}{p}=n$
- It is known that ${(n^p)}^\frac{1}{p}=n$ for any base number $n$, which implies that the $p$-th root of $n^p$ is $n$. Applying this to the sums of $|a_k|^p$ and $|b_k|^p$, we have:

\begin{gather}
\Sigma_{k=1}^n |a_k + b_k|
\leq
\Sigma_{k=1}^n |a_k|
+
\Sigma_{k=1}^n |b_k|
\end{gather}

Step 2: Property of Summation (Addition)

\begin{gather}
\Sigma_{k=1}^n |a_k + b_k|
\leq
\Sigma_{k=1}^n |a_k| + |b_k|
\end{gather}

Step 3: Prove by Triangle Inequality

- Disregarding the Sigma Notation, we get:
\begin{gather}
|a_k + b_k|
\leq
|a_k| + b_k|
\end{gather}

- Since, $(a_k, b_k) > 0$, then $(a + b) = a + b$, and $|a| = a_k$, $|b| = b$:
- Substitute:
\begin{gather}
|a + b| = a + b = |a| + |b|
\end{gather}

**In Conclusion**, the inequality is proven.

{|x| = }

\text{Case 4: }{x < 0, y \geq 0} \\
\end{gather}

\begin{gather}
|x + y| = x + y = |x| + |y|

- Let, $|x + y| = -(x + y)$, and $|x| = -x, |y| = -y$
- Substitution:
\begin{gather}
|x + y| = -(x + y) = -x + (-y) = |x| + |y|
\end{gather}

Case 3: $x \geq 0, y < 0$

- Let, $|x| = x, |y| = -y$
- Substitution:
\begin{gather}

For $x + y \geq 0$:
\begin{gather}
x + y \leq x - y \\  
y \leq -y
\end{gather}

For $x + y < 0$:
\begin{gather}
-(x + y) \leq x - y \\
-x - y \leq x - y \\

\end{gather}}

Case 4: $x < 0, y \geq 0$

- Let, $|x| = -x, |y| = y$
- Substitution:
\begin{gather}

For $x + y < 0$:

**In Conclusion**, in all cases, $|x+y| \leq |x| + |y| \in \mathbb{R} $

4. Prove Sedrakayan's Lemma $\forall u_i, v_i \in \mathbb{R}^+$:
\begin{gather}
\frac{(\Sigma_{i=1}^n u_i)^2}{\Sigma_{i=1}^n v_i}
\leq
\Sigma_{i=1}^n \frac{(u_i)^2}{v_i}
\end{gather}


Step 1: To prove Sedrakayan’s Lemma directly, we need to show that for all $(u_i, v_i \in \mathbb{R}^+)$, the inequality

$$\left(\frac{\sum_{i=1}^n u_i}{\sum_{i=1}^n v_i}\right)^2 \leq \frac{\sum_{i=1}^n (u_i)^2}{v_i}$$

holds true.

Step 2: The **Cauchy-Schwarz Inequality** states that for any real numbers $(a_i)$ and $(b_i)$,

$$\left(\sum_{i=1}^n a_i b_i\right)^2 \leq \left(\sum_{i=1}^n (a_i)^2\right) \left(\sum_{i=1}^n (b_i)^2\right)$$

Step 3: Let $a_i = \sqrt{u_i}$ and $b_i = \frac{1}{\sqrt{v_i}}$. Then, the Cauchy-Schwarz Inequality becomes

$$\left(\sum_{i=1}^n \sqrt{u_i} \cdot \frac{1}{\sqrt{v_i}}\right)^2 \leq \left(\sum_{i=1}^n u_i\right) \left(\sum_{i=1}^n \frac{1}{v_i}\right)$$

**SIMPLIFY:**

Step 4: The *left-hand side* simplifies to

$$\left(\frac{\sum_{i=1}^n u_i}{\sum_{i=1}^n v_i}\right)^2$$

Step 5: The *right-hand side* simplifies to

$$\frac{\sum_{i=1}^n (u_i)^2}{v_i}$$

**In Conclusion**, since the left-hand side is less than or equal to the right-hand side, we have proven Sedrakayan’s Lemma.
