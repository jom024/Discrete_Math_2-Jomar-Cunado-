#### Question 1:

To show that the expected value $E[y] = \sum y \cdot P(y) = n \cdot p_i$, let's calculate $E[Y_i]$, the expected value of $Y_i$.

The expected value of a random variable $Y_i$ is defined as:

$$
E[Y_i] = \sum_{y_i} y_i \cdot P(Y_i = y_i)
$$

Given that $Y_i$ follows a binomial distribution with parameters $n$ and $p_i$, the probability mass function $P(Y_i = y_i)$ is:

$$
P(Y_i = y_i) = \binom{n}{y_i} p_i^{y_i} (1 - p_i)^{n - y_i}
$$

Substituting this into the formula for the expected value:

$$
E[Y_i] = \sum_{y_i} y_i \cdot \binom{n}{y_i} p_i^{y_i} (1 - p_i)^{n - y_i}
$$

This is the expected value of a binomial distribution, which is $n \cdot p_i$. Therefore, $E[Y_i] = n \cdot p_i$.

Consequently, since $Y_1, Y_2, \ldots, Y_k$ are independent random variables, the expected value of their sum is the sum of their expected values:

$$
E[Y] = E[Y_1 + Y_2 + \ldots + Y_k] = E[Y_1] + E[Y_2] + \ldots + E[Y_k]
$$

Substituting $E[Y_i] = n \cdot p_i$ for each $Y_i$:

$$
E[Y] = n \cdot p_1 + n \cdot p_2 + \ldots + n \cdot p_k = n \cdot (p_1 + p_2 + \ldots + p_k) = n \cdot 1 = n
$$

Thus, the expected value $E[Y] = n \cdot p_i$.

The probability distribution of the experiment can be derived as follows:

Since each trial results in either success ($S$) or failure ($F$), and the trials are independent, the probability of observing $y$ successes in $n$ trials follows a binomial distribution.

The probability of observing exactly $y$ successes in $n$ trials is given by the binomial probability formula:

$$[ P(Y=y) = \binom{n}{y} p^y q^{n-y} ]$$

where:
- $p$ is the probability of success on a single trial.
- $q = (1 - p)$ is the probability of failure on a single trial.
- $n$ is the total number of trials.
- $y$ is the number of successes observed.

This formula represents the probability distribution of the experiment.

#### Question 2:

To show that the expectation of this probability distribution is $E[y] = \sum y \cdot P(y) = np$:

The expectation, $E[y]$, of a binomial distribution is given by the formula:

$$[ E[y] = \sum_{y=0}^{n} y \cdot P(Y=y) ]$$

Substituting the binomial probability formula, we get:

$$[ E[y] = \sum_{y=0}^{n} y \cdot \binom{n}{y} p^y q^{n-y} ]$$

This sum simplifies to $ np $, which proves that $E[y] = np$.

#### Question 1:

For a Poisson random variable $(Y)$ with parameter $(\lambda)$, the expected value (mean) $(E[Y])$ can be derived from the definition of the Poisson distribution:

$$[ p(y) = \frac{\lambda^y}{y!} e^{-\lambda}, \quad y = 0, 1, 2, \ldots ]$$

To find the expected value $(E[Y]$), we use the definition of expectation:

$$[ E[Y] = \sum_{y=0}^{\infty} y \cdot p(y) ]$$

Plugging in the probability mass function:

$$[ E[Y] = \sum_{y=0}^{\infty} y \cdot \frac{\lambda^y}{y!} e^{-\lambda} ]$$

We can rewrite $( y \cdot \frac{\lambda^y}{y!} )$ as $( \lambda \cdot \frac{\lambda^{y-1}}{(y-1)!} )$. This shift in the index simplifies the summation. Let $( z = y - 1 )$:

$$[ E[Y] = \sum_{y=0}^{\infty} y \cdot \frac{\lambda^y}{y!} e^{-\lambda} = \lambda e^{-\lambda} \sum_{y=1}^{\infty} \frac{\lambda^{y-1}}{(y-1)!} ]$$

By changing the index of summation from $( y )$ to $( z )$:

$$[ E[Y] = \lambda e^{-\lambda} \sum_{z=0}^{\infty} \frac{\lambda^z}{z!} ]$$

Notice that the sum $( \sum_{z=0}^{\infty} \frac{\lambda^z}{z!} )$ is the Taylor series expansion of $( e^{\lambda} )$:

$$[ \sum_{z=0}^{\infty} \frac{\lambda^z}{z!} = e^{\lambda} ]$$

Thus:

$$[ E[Y] = \lambda e^{-\lambda} \cdot e^{\lambda} = \lambda ]$$

So, the expected value of a Poisson distribution with parameter $(\lambda)$ is:

$$[ \boxed{E[Y] = \lambda} ]$$

#### Question 2:

In this problem, we are given that the mean density of seedlings is $(\lambda = 5)$ seedlings per square yard. A forester randomly locates ten 1-square-yard sampling regions. Our objective is to find the probability that none of these regions will contain seedlings.

Since the presence of seedlings in different regions is independent, we can model the number of seedlings in each 1-square-yard region as a Poisson random variable with $(\lambda = 5)$.

The probability that a single region contains zero seedlings (using the Poisson distribution formula) is:

$$[ P(Y = 0) = \frac{\lambda^0}{0!} e^{-\lambda} = e^{-5} ]$$

Since there are ten independent regions, and we want the probability that none of these ten regions contain seedlings, we raise the single-region probability to the tenth power:

$$[ P(\text{none of the regions contain seedlings}) = (e^{-5})^{10} = e^{-50} ]$$

Thus, the probability that none of the ten 1-square-yard sampling regions will contain seedlings is:

$$[ \boxed{e^{-50}} ]$$
