**GRAPH**

A graph is a fundamental mathematical structure used to represent pairwise relations between nodes, called vertices, with their links called edges.

**DIRECTED GRAPH**

A directed graph is a mathematical structure consisting of nodes, called vertices, connected by arrows, called edges, which have a specific direction from one vertex to another. Each edge has a start vertex and an end vertex, indicating the direction of the relationship.

They can be seen as arrows, an example in the graph below would be, A to D.

```python
import networkx as nx
import matplotlib.pyplot as plt
plt.figure(figsize=(4, 3))
digra = nx.DiGraph()
digra.add_edge("a", "d")
digra.add_edge("b", "a")
digra.add_edge("c", "b")
digra.add_edge("d", "c")
nx.draw(digra, with_labels = True)
plt.title("Directed Graph")
plt.show()
```

**COMPLETE GRAPH**

A complete graph is an undirected graph where each pair of distinct vertices is connected by a unique edge. In other words, every vertex is adjacent to every other vertex.

```python
import networkx as nx
import matplotlib.pyplot as plt
plt.figure(figsize=(4, 3))
comgra = nx.complete_graph(4)
nx.draw_spring(comgra, with_labels = True)
plt.title("Complete Graph")
plt.show()
```

**BIPARTITE GRAPH**

A Bipartite Graph is a graph in which its vertices are divided into two disjoint and independent sets. Every Vertex in either set is not adjacent to the vertex in the same set, however every edge connects vertices from one set to the other

```python
import networkx as nx
import matplotlib.pyplot as plt
plt.figure(figsize=(4, 3))
bigra = nx.Graph()
setA = [1, 2, 3, 4]
setB = [7, 8, 9]
bigra.add_nodes_from(setA, bipartite=0)
bigra.add_nodes_from(setB, bipartite=1)
for nodeA in setA:
    for nodeB in setB:
        bigra.add_edge(nodeA, nodeB)
pos = {node: (0, i) for i, node in enumerate(setA)}
pos.update({node: (1, i) for i, node in enumerate(setB)})
nx.draw(bigra, pos, node_color='skyblue', node_size=500, with_labels=True, edge_color='gray', width=1.5)
plt.title("Bipartite Graph")
plt.show()
```

**HANDSHAKE THEOREM** <br>
In any undirected graph, the theorem states that the sum of the degrees of all vertices is always equal to twice the number of edges in the Graph.*italicized text*

```python
import networkx as nx
import matplotlib.pyplot as plt
plt.figure(figsize=(5, 4))
g1 = nx.complete_graph(5)
g2 = nx.complete_graph(4)
g2 = nx.relabel_nodes(g2, {0: "A", 1: "B", 2: "C", 3: "D", 4: "E"})
exgra = nx.from_edgelist([(4, "X"), ("X", "A")])
G = nx.compose_all([g1, g2, exgra])
nx.draw_spring(G, with_labels = True)
plt.title("Main Graph")
plt.show()
print("Twice number of edges: ")
print(nx.number_of_edges(G) * 2)
degrees = G.degree()
result = sum(deg for node, deg in degrees)
print("Sum of degrees of all vertices: ")
print(result)
```

**Travelling in a Graph**

```python
import networkx as nx
import matplotlib.pyplot as plt
plt.figure(figsize=(5, 4))
g1 = nx.complete_graph(5)
g2 = nx.complete_graph(4)
g2 = nx.relabel_nodes(g2, {0: "A", 1: "B", 2: "C", 3: "D", 4: "E"})
exgra = nx.from_edgelist([(4, "X"), ("X", "A")])
G = nx.compose_all([g1, g2, exgra])
nx.draw_spring(G, with_labels = True)
plt.title("Main Graph")
plt.show()
```

A **Walk** from vertex 'A' to 1 goes via the edge to X, then to 4, 3, 2, and finally arrives at vertex 1. A ‚Üí X ‚Üí 4 ‚Üí 3 ‚Üí 2 ‚Üí 1. <br>
A **Trail** from vertex 'A' to 3 travels through the edge to X, then to 4, and reaches 3. A ‚Üí X ‚Üí 4 ‚Üí 3. Note that a trail does not include any repeated edge. <br>
A **Path** from vertex 'A' to 3 goes through the edge to X, then to 4, and arrives at 3. A ‚Üí X ‚Üí 4 ‚Üí 3. Note that a path does not include any repeated vertex. <br>
A **Closed Walk** from vertex 'A' travels through the edge to X, then to 4, 3, and returns to 'A'. A ‚Üí X ‚Üí 4 ‚Üí 3 ‚Üí A. Note that a closed walk starts and ends at the same vertex. <br>
A **Circuit** from vertex 'B' travels through the edge to 'A', then to 'X', and back to 'B'. B ‚Üí A ‚Üí X ‚Üí B. Note that a circuit is a closed trail. <br>
A **Simple Circuit** from vertex 'X' travels through the edge to 'A', then to 'B', and returns to 'X'. X ‚Üí A ‚Üí B ‚Üí X. Note that a simple circuit is a closed path that starts and ends at the same vertex, without any repeating edges and vertices except for the starting and ending vertex. <br>

**EULER'S CIRCUIT**

2 ‚Üí 4 ‚Üí 3 ‚Üí 2 ‚Üí 0 ‚Üí 1 ‚Üí2

```python
import networkx as nx
import matplotlib.pyplot as plt
plt.figure(figsize=(4, 3))
g = nx.Graph()
g.add_edge(4, 3)
g.add_edge(4, 2)
g.add_edge(3, 2)
g.add_edge(2, 1)
g.add_edge(2, 0)
g.add_edge(1, 0)
nx.draw_spring(g, with_labels = True)
plt.title("Euler's Circuit")
plt.show()
```

**EULER'S TRAIL**

0 ‚Üí 4 ‚Üí 5 ‚Üí 1 ‚Üí 6 ‚Üí 5 ‚Üí 2 ‚Üí 3 ‚Üí 7 ‚Üí 2 ‚Üí 6 ‚Üí 3 ‚Üí 5 ‚Üí 0

```python
import networkx as nx
import matplotlib.pyplot as plt
plt.figure(figsize=(4, 3))
G = nx.Graph()
edges = [(0, 4), (4, 5), (5, 1), (1, 6), (6, 5), (5, 2), (2, 3), (3, 7), (7, 2), (2, 6), (6, 3), (3, 5), (5, 0)]
G.add_edges_from(edges)
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_size=500, node_color='skyblue', font_size=10)
plt.title("Eulerian Trail")
plt.show()
```

**HAMILTONIAN'S CIRCUIT**

0 ‚Üí 1 ‚Üí 2 ‚Üí 3 ‚Üí 0

```python
import networkx as nx
import matplotlib.pyplot as plt
plt.figure(figsize=(4, 3))
G = nx.Graph()
edges = [(0, 1), (1, 2), (2, 3), (3, 0), (0, 2), (1, 3)]
G.add_edges_from(edges)
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True)
plt.title("Hamiltonian Circuit")
plt.show()
```
